```python
import numpy as np
```

```python
f = open('airports.csv', 'r')
```

```python
airport_data = f.read()
```

```python
airport_data = airport_data.split('\n')
```

```python
airport_data = airport_data[1:]
```

```python
airport_data_2 = []
for a in airport_data:
    ad = a.split(',')
    if len(ad) == 7:
        airport_data_2.append(ad)
airport_data= airport_data_2
```

```python
airport_data[0]
```

```python
airport_coordinate_dict={}
for i in range(len(airport_data)):
    airport = airport_data[i][0]
    longitude = airport_data[i][6]
    latitude = airport_data[i][5]
    airport_coordinate_dict[airport] = (longitude, latitude)
```

```python
airport_coordinate_dict['ABE']
```

```python
f.close()
```

```python
f = open('flights.csv', 'r')
```

```python
data = f.read()
```

```python
data = data.split('\n')
```

```python
data = data[1:]
```

```python
data_2 = []
for d in data:
    td = d.split(',')
    if len(td) == 4:
        data_2.append(td)
data= data_2
```

```python
new_data = []
for i, d in enumerate(data):
    temp = []
    origin = d[0]
    dest = d[1]
    origin_coord = airport_coordinate_dict[origin]
    dest_coord = airport_coordinate_dict[dest]
    try:
        temp.append(float(origin_coord[0]))
        temp.append(float(origin_coord[1]))
        temp.append(float(dest_coord[0]))
        temp.append(float(dest_coord[1]))
        temp.append(float(d[2]))
        temp.append(float(d[3]))
        new_data.append(temp)
    except:
        print(d)
```

```python
len(new_data)
```

```python
f.close()
```

```python
arr = np.array(new_data)
```

```python
np.savetxt('preprocessed.csv', arr, delimiter=',') 
```
